data cleaning

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem.snowball import SnowballStemmer
nltk.download('stopwords')
nltk.download('punkt')
```

```python
sn = SnowballStemmer("english")
stop = set(stopwords.words("english"))
```

```python
def clean_text(sms):
  sms = sms.lower()
  sms = re.sub("[^a-z]",' ', sms)
  sms = nltk.word_tokenize(sms)
  sms = [t for t in sms if len(t)>1]
  sms = [sn.stem(word) for word in sms if word not in stop]
  sms = ' '.join(sms)

  return sms
```

```python
clean_text("Get this stuffing for free? playing just Rs.10 t")
```

```python
df['clean_text']= df['text'].apply(clean_text)
```

```python
hamdata = df[df['label']== 'ham']
hamdata = hamdata['clean_text']
```

```python
def wordCloud(data):
  words= ' '.join(data)
  wc = WordCloud(background_color='White')
  wc = wc.generate(words)

  plt.figure(figsize=(10,8))
  plt.imshow(wc)
  plt.axis("off")
  plt.show()
```

```python
wordCloud(hamdata)
```

modelbuilding

```python
from sklearn.model_selection import train_test_split
```

```python
X = cv.fit_transform(df['clean_text']).toarray()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=0)
```

```python
from sklearn.naive_bayes import MultinomialNB

model = MultinomialNB()

model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
pd.DataFrame({'Actual' : y_test, 'predicted': y_pred}).head(10)
```

Evaluation

```python
print("Accuracy on training data :")
print(model.score(X_train,y_train))
```

```python
print("Accuracy on Testing data: ")
print(model.score(X_test,y_test))
```
