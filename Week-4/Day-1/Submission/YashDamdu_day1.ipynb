```python
cols_to_drop = ['url_google_index', 'domain_google_index', 'server_client_domain', 'tld_present_params', 'time_response', 'domain_spf', 'qty_ip_resolved', 'qty_nameservers',
                'qty_mx_servers', 'ttl_hostname', 'url_shortened']

df = df.drop(cols_to_drop, axis = 1)
```

```python
extra_one = ['qty_vowels_domain']
df = df.drop(extra_one, axis = 1)
```

```python
rows, columns = df.shape
print("Number of rows", rows)
print("Number of columns", columns)
```

Perform feature engineering on our dataset

```python
original_features = list(df.columns)
```

```python
original_features
```

```python
data_array = np.array(df)
```

```python
print(data_array)
```

```python
features_indencies = []
attributes = ['url', 'domain', 'file', 'params']
```

```python
new_data = {}
```

```python
for index , name in enumerate(original_features):
  if 'qty' in name and name.split('_')[-1] in attributes:
    features_indencies.append([index, name.split('_'[-1])])
  else:
    new_data[name] = data_array[:,index]
```

```python
for index, attribute in features_indencies:
  if attribute == 'domain':
    if f'qty_char_{attribute}' not in new_data.keys():
      new_data[f'qty_char_{attribute}'] = np.zeros(rows)

    new_data[f'qty_char_{attribute}'] += data_array[:,index]
```

```python
df1 = pd.DataFrame(new_data).astype(int)
```

```python
df1[df1<-1] = -1
df1
```

```python
print(df1.describe())
```

```python
setup(data = df1, target = "phishing")
```

```python
best_model = compare_models()
```

```python
tuned_model = tune_model(best_model, n_iter = 1, optimize = 'F1')
```

```python
predictions = predict_model(tuned_model, data= df1)
```
