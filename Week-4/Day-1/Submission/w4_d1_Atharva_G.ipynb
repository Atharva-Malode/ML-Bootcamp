```python
cols_to_drop = ['url_google_index',
                'domain_google_index',
                'qty_vowels_domain',
                'server_client_domain',
                'tld_present_params',
                'time_response',
                'domain_spf',
                'qty_ip_resolved',
                'qty_nameservers',
                'qty_mx_servers',
                'url_shortened',
                'ttl_hostname'
                ]
df = df.drop(cols_to_drop,axis = 1)
```

```python
rows,columns = df.shape
print("Number of ropes:",rows)
print("Number of columns:",columns)
```

Feature Engineering

```python
original_features = list(df.columns)
```

```python
original_features
```

```python
dataset_array = np.array(df)
```

```python
dataset_array
```

```python
feature_indices = []
attributes = ['url','domain','file','params']
```

```python
new_dataset = {}
```

```python
for index, name in enumerate(original_features):    #index represent index of current and name represents in the current feature
  if 'qty' in name and name.split('_')[-1] in attributes: # 'qty' is present in substring or not
    feature_indices.append([index,name.split('_')[-1]])   # last part is from the attributes
  else:
    new_dataset[name] = dataset_array[:,index]            # it assigns the columns that dont meet the conditions to the new dictionary
```

```python
for index,attribute in feature_indices:     # iterates over the features_indices which contains pairs of index and attributes
  if attribute == 'domain':
    if f'qty_char_{attribute}' not in new_dataset.keys():  # checking if the attributes are present are not in the new_dataset
      new_dataset[f'qty_char_{attribute}'] = np.zeros(rows)# it creates the new key in new_dataset with name ex: qty_char_url
    new_dataset[f'qty_char_{attribute}'] += dataset_array[:,index]
```

```python
df1 = pd.DataFrame(new_dataset).astype(int) # to convert the data into the integer datatypes
```

```python
df1[df1<-1] = -1  #if condition is true i wanna change that element with -1
df1
```

```python
#summary statistics of our dataset

print(df1.describe())
```

```python
#setting
setup (data = df1 ,target = 'phishing') #target variable is phishing which we want to build model
```

```python
#comparing and select the best data model

best_model = compare_models()
```

```python
tuned_model = tune_model(best_model,n_iter = 1, optimize = 'F1')
```

```python
predictions = predict_model(tuned_model,data = df1)
```
